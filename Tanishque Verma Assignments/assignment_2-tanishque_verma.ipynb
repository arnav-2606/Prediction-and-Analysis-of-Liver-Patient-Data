# Handling the missing data

```python
df.isna().sum()
```

Dropping 'deck' column as it contains the most missing values 

```python
df_new = df.drop(columns=['deck'], axis=1)
df_new
```

```python
df_new['age'] = df_new.age.fillna(df_new.age.mean())
```

Dropping the following columns:
1. class
2. who
3. adult_male
4. embark_town
5. alive
They are dropped as the following values are alreaady present in starting columns

```python
df_new = df_new.drop(columns=['class', 'who', 'adult_male', 'embark_town', 'alive'], axis=1)
```

```python
df_new
```
