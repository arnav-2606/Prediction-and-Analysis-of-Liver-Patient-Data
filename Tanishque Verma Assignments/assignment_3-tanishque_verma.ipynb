## Removing outliers

```python
# Taking data till 95th percentile to remove the outliers
df = df[df['price']<=df['price'].quantile(0.95)]
df
```

```python
sns.boxplot(df['price'])
```

## Splitting into dependent and independent variables

```python
# Splitting the data into dependent and independent variables
y = df['price']
y
```

```python
x = df.drop(columns=['price'], axis=True)
x
```

## Encoding the categorical variables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x['mainroad'] = le.fit_transform(x['mainroad'])
x['guestroom'] = le.fit_transform(x['guestroom'])
x['basement'] = le.fit_transform(x['basement'])
x['hotwaterheating'] = le.fit_transform(x['hotwaterheating'])
x['airconditioning'] = le.fit_transform(x['airconditioning'])
x['furnishingstatus'] = le.fit_transform(x['furnishingstatus'])
```

```python
x
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x = scaler.fit_transform(x)
x
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

## Linear Regressor

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
#r2_score
print(r2_score(y_test, y_pred))
```

```python
print(mean_absolute_error(y_test, y_pred))
```

```python
print(mean_squared_error(y_test, y_pred))
```

```python
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```
