```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1. Download the dataset:

# 2. Load the dataset
data = pd.read_csv('titanic.csv')

# 3. Perform Below Visualizations

# Univariate Analysis
sns.countplot(x='survived', data=data)
plt.title('Survival Count')
plt.show()

sns.countplot(x='pclass', data=data)
plt.title('Passenger Class Count')
plt.show()

sns.countplot(x='sex', data=data)
plt.title('Gender Count')
plt.show()

# Bi-Variate Analysis
sns.boxplot(x='survived', y='age', data=data)
plt.title('Survival vs Age')
plt.show()

sns.boxplot(x='pclass', y='fare', data=data)
plt.title('Passenger Class vs Fare')
plt.show()

sns.barplot(x='sex', y='survived', data=data)
plt.title('Survival Rate by Gender')
plt.show()

# Multi-Variate Analysis
sns.scatterplot(x='age', y='fare', hue='survived', data=data)
plt.title('Survival by Age and Fare')
plt.show()

sns.heatmap(data.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()

# 4. Perform descriptive statistics on the dataset
statistics = data.describe()
print(statistics)

# 5. Handle the Missing values
data.dropna(inplace=True)  # Dropping rows with missing values

# 6. Find the outliers and replace the outliers
# Assuming outlier detection and replacement code is implemented separately

# 7. Check for Categorical columns and perform encoding
categorical_columns = ['sex', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone']
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=True)

# 8. Split the data into dependent and independent variables
X = data_encoded.drop('survived', axis=1)
y = data_encoded['survived']

# 9. Scale the independent variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 10. Split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```
